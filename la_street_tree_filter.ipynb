```python
# Latitude, Longitude, Species, DBH
df_selected = df[['Latitude', 'Longitude', 'Species', 'DBH']]
```

## Drop the Vacant Site

```python
# Drop rows where Species is 'Vacant Site (UFD) (Vacant Site)'
df_filtered = df_selected[df_selected['Species'] != 'Vacant Site (UFD) (Vacant Site)']
df_filtered = df_filtered.reset_index(drop=True)
```
